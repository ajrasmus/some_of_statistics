```python
import numpy as np
import matplotlib.pyplot as plt
```

By the work in the pdf, we can define an $\operatorname{Exp}(\beta)$ generator by defining the following functions.

```python
np.random.seed(42)
```

```python
def f(beta, x):
    """The inverse function for the CDF for an Exp(beta) distribution."""
    return -beta * np.log(1-x)
```

```python
def gen_exp(beta):
    """A random number generator from an Exp(beta) distribution."""
    return f(beta, np.random.rand())
```

```python
# A demo of the generator.
gen_exp(3)
```

Below let's confirm empirically that this does give us an actual $\operatorname{Exp}(\beta)$ distribution for $\beta=2$.

```python
beta = 2
```

```python
U = np.random.rand(int(10e6))
```

```python
X = f(beta, U)
```

```python
x = np.arange(0,10,0.1)
y = 1/beta * np.exp(-x / beta)
```

```python
plt.hist(X, bins = np.arange(0,10,0.25), histtype='step', density=True)
plt.plot(x,y, color='black')
plt.title('Plot of PDF for $\operatorname{Exp}(2)$ versus histogram of $F^{-1}(X)$')
plt.show()
```

As you can see, a histogram for $F^{-1}(X)$ approximates the graph of the PDF of $\operatorname{Exp}(\beta)$ quite closely.
