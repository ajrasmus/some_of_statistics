```python
import numpy as np
import matplotlib.pyplot as plt
```

Here are the graphs of $X_n$ versus $n$. The array $X$ has entries $X_n$ for $n$ from $1$ to $10,000$.

```python
np.random.seed(42)
```

```python
ran = np.array(range(1,10001,1))
```

```python
fig, ax = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.5, hspace=0.5)
for i in range(4):
    axis = ax[i // 2][i % 2]
    Y = 2*np.random.randint(0,2,size=10000)-1
    X = np.cumsum(Y)
    axis.plot(ran, X)
    axis.set_title('Graph of $X_n$ versus $n$')
plt.show()
```

The graphs look very different because the variance is equal to $n$. For example, at $n=10000$ we see that $X_n$ varies between $-100$ and $100$ in each of the four cases. This makes sense since $\sqrt{10000}=100$ so $X_{10000}$ is within one standard deviation of the mean in each case.
