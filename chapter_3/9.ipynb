```python
import numpy as np
import matplotlib.pyplot as plt
```

First we plot $\overline{X_n}$ versus $n$ for $X$ a standard normal random variable. The array $X$ below is an array of 10,000 standard normals. The array $Y$ has entry $i$ equal to the sum $X_1 + \ldots + X_n$. The array $Z$ is the array of $\overline{X_n}$'s as $n$ ranges from $1$ to $10,000$.

```python
np.random.seed(42)
```

```python
X = np.random.normal(size=10000)
```

```python
Y = np.cumsum(X)
```

```python
ran = np.array(range(1,10001,1))
```

```python
Z = Y / ran
```

```python
plt.plot(ran, Z)
plt.title('$\overline{X_n}$ versus $n$ for $X$ a standard normal')
```

Now we repeat the process with $X$ a standard Cauchy random variable.

```python
X = np.random.standard_cauchy(size=10000)
```

```python
Y = np.cumsum(X)
```

```python
ran = np.array(range(1,10001,1)) 
```

```python
Z = Y / ran
```

```python
plt.plot(ran, Z)
plt.title('$\overline{X_n}$ versus $n$ for $X$ a Cauchy random variable')
```

We see that the graph in this case has huge jumps and is not asymptotic to any particular value. The reason is that the Cauchy distribution has fat tails. This means you can expect to see huge values of $X_n$ very often. If $X_n$ is huge it will result in a huge change in means from $\overline{X_{n-1}}$ to $\overline{X_n}$.
