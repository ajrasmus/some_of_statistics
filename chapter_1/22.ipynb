```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

We carry out the experiment for $n=10,100,1000$.
Here $y$ is the vector of heads or tails, $c$ is the vector of cumulative sums over time, and $X$ is the total number of heads tossed.

```python
np.random.seed(42)
```

```python
p = 0.3
```

First let's carry out the experiment for $n=10$.

```python
n = 10
x = np.array(range(n))
y = np.random.choice([0,1], size=n, p=[1-p,p])
c = np.cumsum(y)
X = c[-1]

ax = plt.gca()
ax.plot(x, c, color='green', label='Number of heads')
ax.plot(x, p*x, color='blue', label=r'Line $y=px$')
ax.set_xlabel('Tosses')
ax.set_ylabel('Number of heads')
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.title('Number of heads versus number of tosses')
ax.legend()
plt.show()
```

```python
print('Number of heads on {} tosses: {}'.format(n,X))
print('n*p/X: {0:.3f}'.format(n*p/X))
```
