# Part (a)

```python
def cos_basis(n):
    basis = []
    
    for j in range(n):
        if j == 0:
            phi = lambda x: 1
            basis.append(phi)
        else:
            phi = lambda x, k=j: np.sqrt(2) * np.cos(k * np.pi * x)
            basis.append(phi)
    
    return basis
```

```python
def cos_coeffs(X, Y):
    coeffs = []
    n = len(X)
    basis_funcs = cos_basis(n)
    
    for j in range(n):
        phi = basis_funcs[j]
        beta = np.mean(Y * phi(X))
        coeffs.append(beta)
    
    coeffs = np.array(coeffs)
    
    return coeffs
```

```python
def cos_reg(X, Y, J):
    basis = cos_basis(J)
    coeffs = cos_coeffs(X, Y)
    
    def f(x):
        out = [phi(x) for phi in basis]
        return np.dot(coeffs[:J], out)
    
    return f
```

```python
def cos_conf_funcs(X, Y, J, alpha):
    n = len(X)
    basis = cos_basis(n)
    coeffs = cos_coeffs(X, Y)
    r = cos_reg(X, Y, J)
    
    k = n // 4
    sigma = np.sqrt((n / k) * (coeffs**2)[n-k+1:n].sum())
    
    dist = chi
    ch = dist.ppf(1-alpha, df=J)
    
    def a(x):
        out = np.array([phi(x) for phi in basis[:J]])
        return np.sqrt((out**2).sum())
    
    c = lambda x: a(x) * sigma * ch / np.sqrt(n)
    
    l = lambda x: r(x) - c(x)
    u = lambda x: r(x) + c(x)
    
    return l, r, u
```

```python
def plot_cos_conf_band(X, Y, J, alpha, ax):
    xs = np.linspace(0, 1, 1000)
    
    l, r, u = cos_conf_funcs(X, Y, J, alpha)
    
    L = list(map(l, xs))
    R = list(map(r, xs))
    U = list(map(u, xs))
    
    ax.plot(xs, L, c='blue', label='Lower bound')
    ax.plot(xs, R, c='red', label='Regression function')
    ax.plot(xs, U, c='green', label='Upper bound')
    ax.scatter(X, Y, c='black', alpha=0.1)
#     ax.legend()
```

```python
ax = plt.gca()
plot_cos_conf_band(x, Y, 50, 0.05, ax)
```

```python
fig, axs = plt.subplots(5, 2)
fig.subplots_adjust(wspace=0.5, hspace=0.5)

for i in range(1, 11):
    J = 10*i
    ax = axs[(i - 1) // 2][(i - 1) % 2]
    plot_cos_conf_band(x, Y, J, 0.05, ax)
```

# Part (b)

```python
def haar_wave_reg(X, Y):
    n = len(X)
    J = int(np.log2(n)) + 1
    
    alpha = (np.array(list(map(phi, X))) * Y).mean()

    D = np.zeros(2**J - 1)
    for j in range(J):
        for k in range(2**j):
            psi_ = psi(j, k)
            D[2**j + k - 1] = (np.array(list(map(psi_, X))) * Y).mean()
    
    D_sub = D[2**(J-1):]
    sigma = np.sqrt(n) * np.median(np.abs(D_sub)) / 0.6745
    beta = D * (np.abs(D) > sigma * np.sqrt(2 * np.log(n) / n))
    
    psis = [psi(j,k) for j in range(J) for k in range(2**j)]
    
    def f(x):
        wavelet_output = np.array([psi(x) for psi in psis])
        return alpha * phi(x) + np.dot(beta, wavelet_output)
    
    return f
```

```python
def phi(x):
    if 0 <= x and x < 1:
        return 1
    else:
        return 0
```

```python
def psi_mother(x):
    if 0 <= x and x <= 1/2:
        return -1
    elif 1/2 < x <= 1:
        return 1
    else:
        return 0
```

```python
def psi(j, k):
    return lambda x: 2**(j/2) * psi_mother(2**j * x - k)
```

```python
g = haar_wave_reg(x, Y)
```

```python
x_ = np.linspace(0,1,1000)
```

```python
y = list(map(g, x_))
```

```python
plt.plot(x_, y, c='blue')
plt.scatter(x, Y, c='green', alpha=0.5)
```
