```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

We have $\mathrm{Bernoulli}(p)$ random variables $X_1,\ldots, X_n$. Example 1 shows that if we use the prior $p\sim \mathrm{Beta}(\alpha,\beta)$, then the posterior is $p|x^n\sim \mathrm{Beta}(\alpha+s,\beta+n-s)$ where $s=\sum x_i$ is the number of successes.

```python
X = np.array([0,1,0,1,0,0,0,0,0,0])
```

```python
n = len(X)
```

```python
s = X.sum()
```

```python
print(f'n: {n}\ns: {s}')
```

```python
alphas = np.array([1/2, 1, 10, 100])
```

```python
means = np.empty(len(alphas))
```

```python
for i, alpha in enumerate(alphas):
    # Calculate the posterior mean for a
    # Beta(alpha,alpha) prior, for
    # alpha=1/2, 1, 10, 100
    mean = beta.mean(alpha+s, alpha+n-s)
    means[i] = mean
```

Here is the plot of the posterior mean for parameter $\alpha \in \{0.5, 1, 10, 100\}$.

```python
plt.scatter(alphas, means, c='blue', alpha=0.5)
plt.xlabel(r'Parameter $\alpha$ in $\mathrm{Beta}(\alpha, \alpha)$')
plt.ylabel(r'Posterior mean')
plt.title(r'Posterior mean with respect to different priors $\mathrm{Beta}(\alpha, \alpha)$')
```

This seems a little incomplete, so let's plot the posterior mean for $p$ with respect to the prior $\mathrm{Beta}(\alpha,\alpha)$ for many values of $\alpha$.

```python
alphas = np.arange(0,300,0.5)

means = np.empty(len(alphas))

for i, alpha in enumerate(alphas):
    mean = beta.mean(alpha+s, alpha+n-s)
    means[i] = mean
```

```python
plt.plot(alphas, means, c='blue', alpha=0.5)
plt.xlabel(r'Parameter $\alpha$ in $\mathrm{Beta}(\alpha, \alpha)$')
plt.ylabel(r'Posterior mean')
plt.title(r'Posterior mean with respect to different priors $\mathrm{Beta}(\alpha, \alpha)$')
```

We can now see the asymptotic behavior of the posterior mean clearly.
