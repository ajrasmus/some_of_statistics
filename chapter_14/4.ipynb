```python
from numpy.random import normal
from scipy.linalg import sqrtm
import numpy as np
import matplotlib.pyplot as plt
```

A $N(\mu,\Sigma)$ random variable can be obtained by scaling a $N(0,I)$ by $\Sigma^{1/2}$ and adding $\mu$. 

```python
def normal(mu, Sigma, size=1):
    """ Return a N(mu,Sigma) random vector."""
    if size == 1:
        return mu + sqrtm(Sigma) @ np.random.normal(loc=0, scale=1, size=len(mu))
    else:
        return np.array([mu + sqrtm(Sigma) @ np.random.normal(loc=0, scale=1, size=len(mu)) for _ in range(size)])
```

Let's draw a scatter plot to make sure things look right.

```python
np.random.seed(42)
```

```python
X = normal(np.array([1,2]), np.array([[1,1],[1,2]]), size=1000)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.5)
plt.title(r"Scatter plot of a 1000 samples from a $N(\mu,\Sigma)$ distribution")
```
